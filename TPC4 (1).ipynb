### O seguinte trabalho consiste numa aplicação que realiza operações algébricas usando frações criadas pelo utilizador.
### É constituido por diversas funções que tem como função criar, somar, subtrair, multiplicar e dividir frações.
### Contém também um menu inicial com várias opcões, nas quais o utilizador pode escolher a operação que desejar e pode também terminar a aplicação, caso escolha a opcão "sair".

```python

def mdc(n, d):
    res=1
    i=0
    if n>d:
        n, d=d,n
        for i in range(2,n+1):
            if n%i==0 and d%i==0:
                res=i

    elif n<d:
        for i in range(2,n+1):
            if n%i==0 and d%i==0:
                res=i
    
    else:
        res=1
        
    return res
    
def simplificar(f):
    n,d=f
    div=mdc(n,d)
    di=(int(n/div),int(d/div))
    return (di)

def soma(f1,f2):
    n1,d1=f1
    n2,d2=f2
    res=(int(n1*d2+n2*d1),d1*d2)
    som=simplificar(res)
    return som

def subtracao(f1,f2):
    n1,d1=f1
    n2,d2=f2
    sub=((n1*d2-n2*d1),d1*d2)
    res=simplificar(sub)
    return res

def multiplicacao(f1,f2):
    n1,d1=f1
    n2,d2=f2
    mult=(n1*n2,d1*d2)
    res=simplificar(mult)
    return res

def divisao(f1,f2):
    n1,d1=f1
    n2,d2=f2
    div=(n1*d2,n2*d1)
    res=simplificar(div)
    return res


def criarfracao():
    n=int(input("Escreva o numerador da fracao: "))
    d=int(input("Escreva o denominador da fracao:  "))
    f=(n,d)
    return f


print("""Selecione uma opção:
1)Criar as frações
0)Sair""")

opcao1=input("Selecione a opção:  ")
while opcao1!="0":
    if opcao1=="1":

        f1= criarfracao()
        f2= criarfracao()

        print("""1)Somar as frações.
2)Subtrair as frações.
3)Multiplicar as frações.
4)Dividir as frações
0)Sair""")



        opcao=input("Selecione a opção:  ")

        if opcao=="1":
                resposta=soma(f1,f2)
                print(resposta)

        elif opcao=="2":
                resposta=subtracao(f1,f2)
                print(resposta)

        elif opcao=="3":
                resposta=multiplicacao(f1,f2)
                print(resposta)

        elif opcao=="4":
                resposta=divisao(f1,f2)
                print(resposta)

        elif opcao=="0":
                exit(1)

        else:
                print("Escolha uma das opcoes:  ")

else:
            exit(1)
```
