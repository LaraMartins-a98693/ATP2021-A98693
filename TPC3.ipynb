   ###### A função implementada tem como nome __BubbleSort__, e consiste na troca de 2 elementos consecutivos de uma lista até que esta esteja completamente ordenada.
   ###### Para isso, foi criado um ciclo __while__ que percorre no máximo o valor do __quadrado do tamanho__ da lista a ordenar.
   ###### Utilizando o __if__, é possível determinar se cada par de elemento deve ou não trocar de posição, avaliando qual o maior entre eles. Para que estes troquem a posição, são usados de forma consecutiva os métodos __pop__ e __append__. 
   ###### No fim, a função retorna a lista ordenada.
   ###### A função BubbleSort está integrada numa aplicação que consegue criar uma lista manualmente, criar uma lista aleatoriamente, mostrar a lista corrente e apagar a lista.



```python
import random


def criar_listaIn():
    lista=[]
    n=int(input("Introduza o nº de elementos da lista"))
    i=1
    while len(lista)<n:
        elem=input("Introduza o elemento"+str(i)+":")
        lista.append(elem)

    return lista


def criarlistaA():
   n=int(input("Introduza o nº de elementos da lista"))
   i=0
   while len(lista)<n:
       elem=random.randrange(1,200)
       lista.append(elem)
       i=i+1
   return lista


def bubblesort(lista):
    n_orden=1
    i=0
    while n_orden<=len(lista)**2:
        if lista[i]>lista[i+1]:
            num=lista[i]
            num_seguinte=lista[i+1]
            lista.pop(i)
            lista.pop(i)
            lista.insert(i,num)
            lista.insert(i,num_seguinte)
            i=i+1
        elif lista[i]<lista[i+1]:
            i=i+1
            
        n_orden=n_orden+1
        if i==len(lista)-1:
            i=0
        

            
            
    return lista

def menu():
    print("""\n\n\n(1)Gerar uma lista manualmente;
(2)Gerar uma lista de valores aleatórios;
(3)Escrever a lista corrente no stdout;
(4)Limpar a lista;
(5)Ordenar a lista com o BubbleSort;
(0) Sair do programa""")

    
lista=[]

opcao=-1
while opcao!=0:
        menu()
        opcao=input("\nIntroduza uma opção \n")
        if opcao=="1":
            lista=criar_listaIn()
        elif opcao=="2":
            lista=criarlistaA()
        elif opcao=="3":
            print(lista)
        elif opcao=="4":
            lista.clear()
        elif opcao=="5":
            lista=bubblesort(lista)
        elif opcao=="0":
            break
        else:
            print("Opção inválida!")
```
